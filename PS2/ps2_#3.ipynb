```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 1, 101)
print(x)
```

```python
x = np.linspace(0, 1, 101)
n1,n0 = 6,4
y = x**n1 * (1-x)**n0
plt.title("MLE, n=" + str(n1+n0) + ": " + str(n1) + " 1s, " + str(n0) + " 0s") 
plt.xlabel("θ") 
plt.ylabel("L(θ)") 
plt.plot(x, y) 

ymax = max(y)
xmax = x[np.argmax(y)]
plt.axvline(x=xmax, color="red")

plt.show()
print("Max likelihood is " + str(ymax) + " at θ=" + str(xmax))
```
